## Pivot tablolar


```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
help(pd.pivot)
```

```python
df.describe()
```

```python
df
```

```python
pd.pivot(data=df,index="Name",columns="Legendary", values="Generation") # Burada pokemonların isimlerini ve efsanevi mi değil mi 
                                                                        # eğer efsaneyse true sütunun altına kaçınçı nesil olduğu değilse de false sütunun altına yazacak.
```

```python
# Pokemon türlerinin nesillere göre ortalama saldırı değeri nedir ? 
pd.pivot_table(data=df,index="Type 1",columns="Generation",values="Attack")
```

```python
# Türlerin nesillere göre hp değerleri nelerdir ? 
 
pd.pivot_table(data=df, index="Type 1",columns="Generation",values="HP")
```

```python
df[(df["Type 1"] == "Dragon") & (df["Generation"] == 1)].mean(numeric_only=True)     # attack sorusunun sağlamasını yaptık 
```

```python
pd.pivot_table(data=df,index="Generation",aggfunc="sum")
```

```python
yedek = pd.read_csv("pokemon.csv")
yedek = yedek._get_numeric_data()

# artık aggfunc fonksiyonu otomatik numeric only komutunu kaldırdığı için ya elle sayısal olmayan değerleri drop fonkisyonuyla silicez
# yada yukarıdaki yöntemler sadece numeric datayı bir yerde tutacaz
```

```python
pd.pivot_table(data=yedek,index="Generation",aggfunc="min")
```

```python
df
```

```python
# Mesela vermesini istediğimiz bir kaç değer varsa ne yapabiliriz
pd.pivot_table(data=df, index="Type 1",values=["Attack","Defense","Sp. Atk"],aggfunc= "min")
```

```python
pd.pivot_table(data=df, index="Type 1", values=["Generation","Speed"])
```

```python
df.groupby("Type 1")[["Generation","Speed"]].mean()
```

```python
# içinde non değerler olan pivot tablolara ne yapabiliriz 

pd.pivot_table(data=df, index= "Generation", columns="Type 1", values="Attack") 
```

```python
round(pd.pivot_table(data=df, index= "Generation", columns="Type 1", values="Attack",fill_value=0),2)
```
