## .apply & np.vectorize

```python
"""
.apply bir fonksiyonu, dataframein içinde çalıştırmamızı sağlıyor.
"""
```

```python
def harf(a):                        #lst[start slice : end slice : step]
    return str(a)[:2]
```

```python
df["Harf2"] = df["Name"].apply(harf)
df
```

```python
df = df.drop("Harf2", axis=1)
```

```python
df
```

```python
df["Total"].describe()
```

```python
def Powerful(Total):
    if Total < 325:
        return "*"
    elif Total < 430:
        return "**"
    elif Total < 500:
        return "***"
    elif Total < 700 and Total > 500:
        return "****"
    else:
        return "*****"
```

```python
df["Powerful"] = df["Total"].apply(Powerful)
df
```

```python
def def_pwr(Defense,Sp_def):
    if Defense + Sp_def > 240:
        return "Shield"
    else:
        return "Neutral"
```

```python
df["Defender"] = df[["Defense", "Sp. Def"]].apply(lambda df: def_pwr(df["Defense"], df["Sp. Def"]),axis=1)
```

```python
df[df["Defender"] == "Shield"]
```

```python
import numpy as np
```

```python
def atck_pwr(Atk,sp_atk):
    if Atk + sp_atk > 240:
        return "Sword"
    else:
        return "Neutral"
```

```python
df["Attacker"] = np.vectorize(atck_pwr)(df["Attack"],df["Sp. Atk"])
```

## APPLY KOMUTU MU YOKSA VECTORİZE KOMUTU MU DAHA İYİ

```python
import timeit
```

```python
setup = """
import numpy as np
import pandas as pd

df = pd.read_csv("pokemon.csv")

def def_pwr(Defense,Sp_def):
    if Defense + Sp_def > 240:
        return "Shield"
    else:
        return "Neutral"


"""

stmt_one = """
df["Defender"] = df[["Defense", "Sp. Def"]].apply(lambda df: def_pwr(df["Defense"], df["Sp. Def"]),axis=1)

"""

stmt_two = """
df["Defender"] = np.vectorize(def_pwr)(df["Defense"], df["Sp. Def"])
"""

```

```python
timeit.timeit(setup=setup,
              stmt=stmt_one,
              number=100)
```

```python
timeit.timeit(setup=setup,
              stmt=stmt_two,
              number=100)
```
