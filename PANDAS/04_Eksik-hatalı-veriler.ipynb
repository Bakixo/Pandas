## ULTRA İYİ BİR YÖNTEM DAHA

```python
""" transform metodu : 
transform metodu, genellikle gruplama işlemleri sonrasında, gruplara özgü dönüşümler uygulamak için kullanılır.
Örneğin, bir DataFrame'i bir kategorik sütuna göre grupladıktan sonra, her bir grup için ayrı ayrı 
ortalama merkezleme yapmak gibi.
"""
```

```python
df["mass"].fillna(df["mass"].mean()) # bunu yazdığımızda mass i ortalamalarıyla dolduruyor ancak kıyaslama tam olarak doğru olmıyor
                                     # çünkü yoda ile insanın kıyaslandığını düşünün tabiki insan daha ağır geliyor  
```

```python
df
```

```python
df[["name","sex","mass"]]
```

```python
df["mass"] = df["mass"].fillna(df.groupby("sex")["mass"].transform("mean"))
```

```python
df[["name","sex","mass"]] #reyin kilosu non iken artık 47 olarak belirlendi nasıl yaptı ?
                          # female olanları aldı leia padme gibi ikisin aldı ve arasında reyin kaç kilo olabileceğine göre yorum yaptı
```

```python
df[["height","sex","species","name"]]
```

```python
df["height"] = df["height"].fillna(df.groupby(["sex","species"])["height"].transform("mean"))
```
