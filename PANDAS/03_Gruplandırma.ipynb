## .groupby()

* .mean(): Grubun ortalaması
* .sum(): Grubun toplamı
* .size(): Grup boyutu
* .count(): Grup sayısı
* .min(): min grup değeri
* .max(): max grup değeri
* .describe(): Aynısı
* .std(): Standart sapma
* .var(): Varyans
* .sem(): Standard error of the mean
* .first(): Gruptaki ilk değer
* .last(): Gruptaki son değer


```python
df.groupby("cyl")
```

```python
df.groupby("cyl").mean(True) #numeric only var daha öncede hepsine uygulamamız gerektiğini söylemiştik bu yüzden içine True geliyor
```

```python
cyl_ort = df.groupby("cyl").mean(True)
```

```python
df.groupby(["cyl","vs"]).mean(True)
```

## Aggregate

```python
help(df.agg)
```

```python
# df.agg(["min","max","std","mean"])
```

```python
# df.agg(["mean","max","min"])[["mpg","cyl","gear"]]
```

```python
df.agg({
    "hp" : ["max","min"],
    "cyl" : ["mean"],
    "gear" : ["mean", "min","max"]
})
```

```python
df.groupby("cyl").agg({
    "hp" : ["max","min","mean"],
    "gear" : ["min","max"]

})
```
